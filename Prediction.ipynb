```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2

model = tf.keras.models.load_model('/content/drive/MyDrive/Flower_Recognition/Flower Recognizer.hdf5')
```

```python
flowers = ['daisy','sunflower','rose','dandelion','tulip']
img =  plt.imread('/content/download.jpg',1)
plt.imshow(img)
x = cv2.resize(img,(48,48))
x = np.expand_dims(x,axis=0)
x = x/255
y = model.predict(x)
y = np.argmax(y)
print(flowers[y])
```
