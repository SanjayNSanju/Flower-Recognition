```python
import tensorflow as tf
import numpy as np
import cv2

model = tf.keras.models.load_model('/content/drive/MyDrive/Flower_Recognition/Flower Recognizer.hdf5')
flowers = ['daisy','sunflower','rose','dandelion','tulip']

img =  cv2.imread('/content/t.jpg',1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
x = cv2.resize(img,(48,48))
x = np.expand_dims(x,axis=0)
x = x/255
y = model.predict(x)
y = np.argmax(y)
print(flowers[y])
```
